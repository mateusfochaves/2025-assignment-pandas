```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def load_data():
    """Load data from the CSV files referendum/regions/departments."""
    
    data = pd.read_csv("data/referendum.csv", index_col=0, parse_dates=True)
    referendum = pd.DataFrame(data)

    data = pd.read_csv("data/regions.csv", index_col=0, parse_dates=True)
    regions = pd.DataFrame(data)

    data = pd.read_csv("data/departments.csv", index_col=0, parse_dates=True)
    departments = pd.DataFrame(data)
    
    return referendum, regions, departments
```

```python
ref, reg, dep = load_data()
```
